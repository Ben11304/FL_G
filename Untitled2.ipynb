```python
import pandas as pd
import numpy as np
import data_work

data=pd.read_csv("processed_20k.csv")
```

```python
data_work.count_label_data(data["subcategory "])
```

```python
from sklearn.model_selection import train_test_split 
import torch
traindata,test_data=train_test_split(data,test_size=0.4,random_state=42,stratify=data['subcategory '])
data = traindata.reset_index(drop=True)
test_data = test_data.reset_index(drop=True)
client_data,server_data=train_test_split(traindata,test_size=0.01,random_state=42)
client_data = client_data.reset_index(drop=True)
server_data = server_data.reset_index(drop=True)
trainloaders,testloaders=data_work.data_split(client_data,"subcategory ", 4,0.1,False)
```

```python
server_data.to_csv("./data/server_data.csv")
test_data.to_csv("./data/test_data.csv")
for i in range(4):
    trainloaders[i].to_csv(f"./data/trainloader_{i}.csv")
    testloaders[i].to_csv(f"./data/testloader_{i}.csv")
```
